# Bootstrap in Python


A student sent me there bootstapping homework in python and I thought I should give it a try too.

```python
import random
```

```python
data = []

for i in range(23):
  data.append(1)
for i in range(7):
  data.append(0)

data
```

To create one bootstrap sample I randomly select 30 items from the list.

```python
boot = random.choices(data, k = len(data))

boot
```

To do the bootstrap distribution, I'll repeat that process multiple times.

```python
def mean(list):
  return sum(list)/len(list)

mean(boot)
```

```python
bootdist = []

for i in range(1000):
  bootdist.append(mean(random.choices(data, k = len(data))))

bootdist
```

Then the center aught to be very close to the point estimate

```python
mean(bootdist)
```

And here is a histogram of the distribution:

```python
import matplotlib.pyplot as plt

plt.hist(bootdist)
plt.show()
```
